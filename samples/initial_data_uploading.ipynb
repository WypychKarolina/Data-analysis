# libraries importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# sorting csv files out

```python
population_initial = pd.read_excel("/home/wypychkarola/mentoring/Data-analysis/res/population_v.xlsx", header = 2) 

population = population_initial[:16]
del population["Dane"]
del population["nts"]
```

```python
population_T = population.transpose()

population_T.columns = population_T.iloc[0] 
popul = population_T.drop(axis=0, labels = "nazwa")

popul.drop(popul[(popul.index > 2021) == True].index, inplace = True)
popul.drop(popul[(popul.index < 2006) == True].index, inplace = True)
popul = popul.reset_index()

popul.to_excel("/home/wypychkarola/mentoring/Data-analysis/res/population_ok.xlsx")
```

```python
colnames = list(popul.columns)

popul.plot(x="index", y=list(popul.columns[1:]), kind = 'line', legend=True, 
                 subplots = True, sharex = True, sharey = False, figsize = (20,10), 
                 ls="none", marker="o", layout = (4,4), label = colnames[1:])
plt.show()
```
